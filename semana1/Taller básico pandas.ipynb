# Taller b√°sico pandas - numpy

```python
import pandas as pd
import numpy as np
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos

cancion = pd.Series(['ojoschiquiticos', 'latraicion', 'laex'])
artista = pd.Series(['omar','wisinyandel', 'fernando'])
duracion = pd.Series([5,3,8])

df_weather = pd.DataFrame ({'cancion': cancion, 'artista': artista, 'duracion':duracion})
df_weather
```

```python
# Imprimir la cola del dataframe
print(df_weather.tail())
```

```python
# Acceder a la columna duracion
df_weather.loc[:,['duracion']]
```

```python
# Calcular el promedio de las duraciones
promedio_duracion = df_weather['duracion'].mean()

print(f"El promedio de las duraciones es: {promedio_duracion}")
```

```python
# Obtener los nombres de las canciones que tengan duracion mayor que el promedio


canciones_mayor_promedio = df_weather[df_weather['duracion'] > promedio_duracion]
nombres_canciones = canciones_mayor_promedio['cancion'].tolist()

print("Canciones con duraci√≥n mayor que el promedio:", nombres_canciones)
```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor
df_weather['mayor a 3 minutos'] = df_weather['duracion'] > 3

print(df_weather)
```

```python
# Convertir el dataframe a numpy array

array_numpy = df_weather.to_numpy()

print(array_numpy)
```

```python
# Imprimir el shape del array
print(array_numpy.shape)
```

```python
# Hacer un reshape para convertir el arreglo de 3x3 a un arreglo de 12 elementos

array_numpy = df_weather.to_numpy()

# Imprimir la forma original del array
print("Forma original del array:", array_numpy.shape)

array_reshaped = array_numpy.reshape(-1)

# Imprimir el nuevo array y su forma
print("Array redimensionado:", array_reshaped)
print("Forma del array redimensionado:", array_reshaped.shape)
```
